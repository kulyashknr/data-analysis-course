```python
import os
import json
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
from datetime import datetime
from tqdm import tqdm
tqdm.pandas()
%matplotlib inline
import math
```

```python
df = pd.read_json("./test.json")
```

```python
def preprocess_date(text):
    text = str(text)
    if len(text)>len("11 Mar 2022"):
        cre_date = re.compile(r"([0-9]{1,2}[ ]{1,3}(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[ ]{1,3}[0-9]{4})")
        match = re.search(cre_date,text) 
        if len(match.groups())>0:
            return datetime.strptime(match.groups()[0].strip(), '%d %b %Y')
    else:
        return datetime.strptime(text, '%d %b %Y')

def preprocess_categories(cats):
    new_cats = []
    for cat in cats:
        new_cats.append(cat.strip())
    print(new_cats)
    return new_cats

def separate_categories(data):
    for row in data.index:
        if len(data["clean_category"][row]) > 1:
            for category in data["clean_category"][row][1:]:               
                data = data.append({'url': data["url"][row], 'date_publ': data["date_publ"][row], 'title': data["title"][row], 'month': data["month"][row], 'year': data["year"][row], 'clean_category': category}, ignore_index=True)
        data["clean_category"][row] = str(data["clean_category"][row][0])
    return data
```

```python
df["date_format"] = df["date_publ"].progress_apply(preprocess_date)
df["month"]=df["date_format"].apply(lambda x: x.month)
df["year"]=df["date_format"].apply(lambda x: x.year)
df["clean_category"]=df["categories"].progress_apply(preprocess_categories)
df = separate_categories(df)
```
